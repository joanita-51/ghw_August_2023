K-Means Clustering Algorithm: Tutorial

```python
#first, we import the necessary Libraries
import pandas as pd #great for data tables
import numpy as np #great for manipulating data
import matplotlib.pyplot as plt # great for plotting the data and/or visualizing import
from sklearn.cluster import KMeans
```

A Standard K-means Clustering Example
(An article to check out is https://www.r-bloggers.com/2017/08/exploring-assumptions-of-k-means-clustering-using-r/)

```python
df = pd.read_csv('data/faithful.csv')
df
```

```python
plt.scatter(df['eruption'], df['waiting'])
plt.title('Faithful Eruption Dataset')
plt.xlabel('Eruption Length (minutes)')
plt.ylabel('Waiting Time b/w Eruptions (minutes)')
```

```python
# formatting the data to be useful to the KMeans model
X = df.values
X # each list within the array contains the ordered features for a given record
```

```python
# we now build the KMeans model
kmeans = KMeans(n_clusters=2, # specifies the number of clusters (can be based on intuition)
                init = 'random', # specifies how we want to initialize the centroids
                max_iter = 300, # want to specify the maximum number of iterations we will allow for kmeans
                )
kmeans.fit(X) # 'fit' method will fit the kmeans model to the faithful dataset
```

```python
# Results of our fit 
print('Cluster Centers:', kmeans.cluster_centers_[0], kmeans.cluster_centers_[1])
print('SSE:', kmeans.inertia_)
print('# of iterations:', kmeans.n_iter_)
```

```python
output = kmeans.labels_ # shows the classifications that resulted from the model fit
output
```

```python
category0 = df[output == 0]
category1 = df[output ==1 ]
plt.scatter(category0['eruption'], category0['waiting'], label = 'Category O Eruption')
plt.scatter(category1['eruption'], category1['waiting'], label = 'Category 1 Eruption')
plt.title("Faithful Erruption Dataset")
plt.legend(loc= 'upper left')
plt.xlabel("Erruption length (minutes)")
plt.ylabel("Waiting time between Erruptions (minutes)")
```

This Eruption lasts for 30 minutes and came 75.2 minutes after the previous eruption 

```python
category0 = df[output == 0]
category1 = df[output ==1 ]
plt.scatter(category0['eruption'], category0['waiting'], label = 'Category O Eruption')
plt.scatter(category1['eruption'], category1['waiting'], label = 'Category 1 Eruption')
plt.scatter(3, 75.2, label='New Eruption')
plt.title("Faithful Erruption Dataset")
plt.legend(loc= 'upper left')
plt.xlabel("Erruption length (minutes)")
plt.ylabel("Waiting time between Erruptions (minutes)")
```

```python
#Predicting the category to which the new eruption belongs
kmeans.predict([[3, 75.2]])
```

What if the data is non linear? Using a non-sperical clustering example

```python
# Here we use the bull's eye we are only taking columns at x and y
df= pd.read_csv('data/bulls_eye.csv')[['x', 'y']]
df
```

```python
plt.scatter(df['x'], df['y'])
plt.title ("Bull's Eye Hits")
```

```python
kmeans = KMeans(n_clusters = 3, init = 'random', max_iter = 300, )  
kmeans.fit(df.values)

flags = kmeans.labels_
category0 = df[flags == 0]
category1 = df[flags == 1]
category2 = df[flags == 2]

#plotting the categories
plt.scatter(category0['x'], category0['y'], label = 'Category 0')
plt.scatter(category1['x'], category1['y'], label = 'Category 1')
plt.scatter(category2['x'], category2['y'], label = 'Category 2')
plt.legend(loc = "upper left")
plt.title ("Bull's Eye Hits")
```

```python
# Polar Coordinates Transformation
df['r'] = (df['x']**2 + df['y']**2)**0.5 # r^2 = x^2 + y^2
df['theta'] = np.arcsin(df['y'] / df['r']) # theta = arcsin(y/r)
```

```python
plt.scatter(df['r'], df['theta'])
plt.title('Bulls Eye (Polar Coordinates)')
plt.xlabel('radius')
plt.ylabel('theta')
```
